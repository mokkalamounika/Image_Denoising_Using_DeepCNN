```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.manifold import Isomap
```

```python
def display_2d_component_names(model,selected,dataobj):
  colors=["#476A2A",'#785188','#BD3430','#4A2D4E','#875525','#A83683','#4E655E',
          '#3A3120','#535D8E']
  plt.figure(figsize=(14,14))
  plt.xlim(model[:,0].min(),model[:,0].max()+1)
  plt.ylim(model[:,1].min(),model[:,1].max()+1)
  for i in range(len(dataobj)):
    cindex=dataobj[i]%len(selected)
    if dataobj[i] not in selected:
      continue 
    plt.text(model[i,0],model[i,1],str(dataobj[i]),
             color=colors[cindex%9],fontdict={'weight':'bold','size':9})
  plt.xlabel("model feature 0")
  plt.ylabel("model feature 1")
```

```python
pca=PCA(n_components=2)
pca.fit(X)
digits_pca=pca.transform(X)
display_2d_component_names(digits_pca,(0,1),y)
```

```python
tsne=TSNE(random_state=42)
digits_tsne=tsne.fit_transform(X)
```

```python
display_2d_component_names(digits_tsne,(0,1),y)
```

```python
iso=Isomap(n_neighbors=10,n_components=2)
digits_iso=iso.fit(X).transform(X)
```

```python
display_2d_component_names(digits_iso,(0,1),y)
```

```python
def plot_pca_accuracy(model,train,test,ytrain,ytest):
  components=[]
  scores=[]
  for c in range(1,15,1):
    pca=PCA(n_components=c,whiten=True,random_state=0).fit(train)
    Xtrain=pca.transform(train)
    Xtest=pca.transform(test)
    model.fit(Xtrain,ytrain)
    scores.append(model.score(Xtest,ytest))
    components.append(c)
  plt.plot(components,scores,label='train accuracy')
  plt.ylabel("Accuracy")
  plt.xlabel('components')
```

```python
plot_pca_accuracy(svm.LinearSVC(C=10.0),X_train,X_test,y_train,y_test)
```
